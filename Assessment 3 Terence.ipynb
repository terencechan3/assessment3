```python
df = pd.read_csv('openrice.csv')

df.head(10)
```

```python
def pricecat(df):
    if(df['price_range']=="Below $50"):
        return 'Cheap'
    elif (df['price_range']=="$51-100"):
        return 'Not Cheap'
    elif (df['price_range']=="$101-200"):
        return 'Expensive'
    elif (df['price_range']=="$201-400"):
        return 'Very Expensive'
```

```python
df['Price_Category']=df.apply(pricecat,axis=1)
```
